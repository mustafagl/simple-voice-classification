```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# CSV dosyasını oku
df = pd.read_csv('/kaggle/input/top5000-movies/tmdb_5000_movies.csv')
import json

# 'genres' sütunundaki her bir değeri liste olarak değiştir
#df['genres'] = df['genres'].apply(lambda x: [genre["name"] for genre in json.loads(x)])

# Değişikliği göstermek için ilk beş satırı yazdır

# NaN değerleri boş dize ('') ile doldur
df = df.fillna('')

# TF-IDF vektörleme
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['genres'] + ' ' + df['keywords'] + df["original_title"])

# Cosine similarity matrisi
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Öneri yapmak için fonksiyon
def get_recommendations(title, cosine_sim=cosine_sim):
    # Film başlığına göre index'i al
    idx = df[df['title'] == title].index[0]
    # Tüm filmlerin benzerlik skorunu al
    sim_scores = list(enumerate(cosine_sim[idx]))
    # Benzerlik skorlarına göre sırala
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # En benzer 10 filmi al
    sim_scores = sim_scores[1:11]
    # Benzer filmlerin index'lerini al
    movie_indices = [i[0] for i in sim_scores]
    # Benzer filmlerin başlıklarını ve benzerlik skorlarını döndür
    return df['title'].iloc[movie_indices], [i[1] for i in sim_scores]

# Örnek kullanım
movie_title = "The Godfather"
recommendations, scores = get_recommendations(movie_title)
for movie, score in zip(recommendations, scores):
    print(f'{movie}: {score}')
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# CSV dosyasını oku
df = pd.read_csv('/kaggle/input/top5000-movies/tmdb_5000_movies.csv')

# Tarih verisini uygun şekilde işle
df['release_date'] = pd.to_datetime(df['release_date'])
df['release_year'] = df['release_date'].dt.year
df['release_month'] = df['release_date'].dt.month
df['release_day'] = df['release_date'].dt.day
# Sayısal verileri normalize et
scaler = MinMaxScaler()
numerical_columns = ['budget', 'revenue', 'popularity', 'runtime', 'vote_average', 'vote_count', 'release_year','release_month','release_day']
for numeric_value in numerical_columns:
    print(numeric_value)
    df[numeric_value] = df[numeric_value].astype(float).fillna(df[numeric_value].astype(float).mean())

df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Metin tabanlı sütunlar için TF-IDF vektörleme
text_columns = ['genres', 'keywords', 'overview', 'tagline','original_title','production_companies','production_countries','spoken_languages','title']
df = df.fillna('')

text_data = df[text_columns].apply(lambda x: ' '.join(x), axis=1)
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(text_data)

# Kategorik verileri dönüştür
df = pd.get_dummies(df, columns=['original_language'])



tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
tfidf_df=tfidf_df.fillna(0)
# Tüm verileri birleştir
features = pd.concat([df[numerical_columns],
                      df.iloc[:, -1], tfidf_df], axis=1)

# Cosine similarity matrisi

cosine_sim = cosine_similarity(features, features)

# Öneri yapmak için fonksiyon
def get_recommendations(title, cosine_sim=cosine_sim):
    # Film başlığına göre index'i al
    idx = df[df['title'] == title].index[0]
    # Tüm filmlerin benzerlik skorunu al
    sim_scores = list(enumerate(cosine_sim[idx]))
    # Benzerlik skorlarına göre sırala
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # En benzer 10 filmi al
    sim_scores = sim_scores[1:11]
    # Benzer filmlerin index'lerini al
    movie_indices = [i[0] for i in sim_scores]
    # Benzer filmlerin başlıklarını ve benzerlik skorlarını döndür
    return df['title'].iloc[movie_indices], [i[1] for i in sim_scores]

# Örnek kullanım
movie_title = 'The Dark Knight Rises'
recommendations, scores = get_recommendations(movie_title)
for movie, score in zip(recommendations, scores):
    print(f'{movie}: {score}')
```
